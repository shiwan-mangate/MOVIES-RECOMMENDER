```python
## genres,id,keywords,title,overview,cast,crew

movies = movies[['movie_id','genres','keywords','title','overview','cast','crew']]
```

```python
import ast
def convert(obj):
    list = []
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0].cast
```

```python
def convert3(obj):
    list = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3 :
              list.append(i['name'])
        else:
             break
    return list
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    list = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            list.append(i['name'])
            break
    return list
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['crew']+ movies['genres']+movies['keywords']+movies['cast']
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

```python
new_df.head()
```

```python
from nltk.corpus import stopwords
sw_list = stopwords.words("english")

new_df['tags'] = new_df['tags'].apply(lambda x : [item for item in x.split() if item not in sw_list]).apply(lambda x :" ".join(x) )
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
from nltk.stem import SnowballStemmer

sn = SnowballStemmer("english")
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(sn.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
import gensim
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
story = []
for doc in new_df['tags']:
    raw_sent = sent_tokenize(doc)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    window= 10,
    min_count=2,
    vector_size=300,
    sg = 1,
    epochs=20,
    negative=10,
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
len(model.wv.index_to_key)
```

```python
def document_vector(doc):
    doc = [word for word in doc.split() if word in model.wv.index_to_key]
    return np.mean(model.wv[doc],axis = 0)
```

```python
document_vector(new_df['tags'].values[0])
```

```python
vectors = new_df['tags'].apply(document_vector).to_list()
```

```python
vectors = np.array(vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse= True , key= lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse= True , key= lambda x: x[1])[1:6]

    for i in movies_list:  
        print(new_df.iloc[i[0]].title)
```

```python
new_df['title']
```

```python
recommend("El Mariachi")
```
